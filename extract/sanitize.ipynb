```python
import  json
import pandas as pd
```

```python
with open("data.json") as fh:
    data = json.load(fh)
```

```python
for key in data:
    data[key]["description"]  =data[key]["description"].strip()

with open("final.json","w") as fh:
    json.dump(data,fh,indent=4)
```

```python
def data_to_df(data):
    l = []
    for key,value in data.items():
        value["url"] = key
        l.append(value)

    return pd.DataFrame(l)

df = data_to_df(data)
df.head()
```

```python
df["description"] = df["description"].str.strip()
```

```python
df.sample(4)
```

```python
df.isna().sum(axis=0)
```

```python
from langdetect import detect

def is_not_english(text):
    try:
        return detect(text) != 'en'
    except:
        return False 
    
translate_df = df[df.description.apply(is_not_english)==True]
print(translate_df.shape[0])
translate_df.sample(4)
```

```python
new = translate_df["title"].str.split("-").apply(lambda x: x[1] if len(x) == 2 else pd.NA)
new.isna().sum()
```

```python
translate_df[new.isna()]
```

```python
df.loc[translate_df.index,"title"] = new
```

```python
from googletrans import Translator
translator = Translator()
```

```python
def translate(textdf):
    return translator.translate(textdf, dest='en')
```
